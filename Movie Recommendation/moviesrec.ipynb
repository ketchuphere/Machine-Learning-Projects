```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('/bestmovies.csv')
```

```python
movies.info()
```

```python
movies['tags'] = movies['genre'] + movies['overview']
```

```python
movies.head()
```

```python
newdf = movies[['id','title','genre','overview','tags']]
```

```python
newdf=newdf.drop(columns=['genre','overview'])
```

```python
newdf.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vc = cv.fit_transform(newdf['tags'].values.astype('U')).toarray()
```

```python
vc
```

```python
vc.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vc)
```

```python
sim
```
